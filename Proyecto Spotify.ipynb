```python
df_spoty1['Release Date'] = df_spoty1['Release Date'].astype('datetime64[ns]')
df_spoty1.dtypes
```

```python
valores = df_spoty1.loc[:,['Release Date','Spotify Streams']]
valores
valores.sort_values(by='Release Date', ascending=False)
```

```python
valores.loc[valores['Release Date'].dt.year == 2023]
```

```python
valores['Spotify Streams'].sum()
```

```python
my_list = list(range(1991, 2025))
sumax = []
df_year = pd.DataFrame(my_list, columns=['year'])
df_year
for i in my_list:
  valori= valores.loc[valores['Release Date'].dt.year == i]
  sumi = valori['Spotify Streams'].sum()
  sumax.append(sumi)
df_year['streams'] = sumax
df_year.plot.bar(x='year',y='streams')
```
